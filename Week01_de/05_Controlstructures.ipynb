# Kontrollstrukturen
In einem Programm müssen Entscheidungen getroffen und Bedingungen geprüft werden.
Wenn du einen zufälligen Bitgenerator programmieren willst, kannst du den folgenden Code verwenden, der das Vorzeichen einer Zufallszahl prüft:

```python
import numpy as np

# single random number with gaussian distribution
x = np.random.randn(1)
if x < 0.0:
    result = 1
else:
    result = 0
    
print('the random number is: ', x)
print('the resulting random bit is: ', result)
```

Mit elif können weitere Entscheidungen getroffen werden:

```python
# random number with uniform distribuion in the range 0..1
x = np.random.rand(1)
if x < 0.1:
    result = 'x is smaller than 0.1'
elif x < 0.5:
    result = 'x is in the range 0.1..0.5'
else:
    result = 'x is greater than 0.5'
    
print('the random number is: ', x)
print(result)
```

Mit verschiedenen Einrückungsebenen lassen sich Entscheidungspfade abbilden:

```python
# generate two random bits:
x1 = np.random.randn(1) < 0.0
x2 = np.random.randn(1) < 0.0

# evaluate the xor function:
if x1:
    if x2:
        result = False
    else:
        result = True
else:
    if x2:
        result = True
    else:
        result = False

print('x1 = ', x1)
print('x2 = ', x2)
print('x1 xor x2 = ', result)
```

Die if-Klausel kann verschiedene boolesche Ausdrücke kombinieren, z. B. and, or, ...:

```python
# generate two random bits:
x1 = np.random.randn(1) < 0.0
x2 = np.random.randn(1) < 0.0

# evaluate the xor-function in another way:
if x1 and x2:
    result = False
elif (not x1) and (not x2):
    result = False
else:
    result = True
    
print('x1 = ', x1)
print('x2 = ', x2)
print('x1 xor x2 = ', result)
```

Der folgende Code kombiniert Schleifen und Kontrollstrukturen, um Primzahlen zu finden:

```python
for n in range(2, 100):
    IsPrime = True
    Limit = n**0.5
    m = 2
    while IsPrime and (m <= Limit):
        if (n % m) == 0:
            IsPrime = False 
        else:
            m += 1       
    if IsPrime:
        print(n, ' is a prime numer.')
```
