# Public, Protected und Private
Klassenvariablen können privat, geschützt oder öffentlich sein.
Private Variablen erhalten zwei führende Unterstriche.
Geschützte Variablen beginnen mit einem Unterstrich.
Öffentliche Variablen haben keinen Unterstrich.
Im Folgenden ist Radius privat, Height geschützt und Weight öffentlich.

```python
class CGeometricalObject(object):
    
    def __init__(self):
        self.__Radius = 1.0
        self._Height = 2.0
        self.Weight = 3.0
    
    def GetRadius(self):
        return self.__Radius
    
    def GetHeight(self):
        return self._Height
    
    def GetWeight(self):
        return self.Weight
    
AGeometricalObject = CGeometricalObject()
print('The radius of the object is: ', AGeometricalObject.GetRadius())
print('The height of the object is: ', AGeometricalObject.GetHeight())
print('The weight of the object is: ', AGeometricalObject.GetWeight())
```

Auf eine private Variable kann von außen nicht zugegriffen werden:

```python
print('trying direct access to a private variable: ', AGeometricalObject.__Radius)
```

Auf eine öffentliche Variable kann von außen zugegriffen werden.
Solange du dich im gleichen Skript befindest (wie hier), ist auch ein Zugriff auf eine geschützte Variable möglich.

```python
print('trying direct access to a protected variable: ', AGeometricalObject._Height)
print('trying direct access to a public variable: ', AGeometricalObject.Weight)
```
