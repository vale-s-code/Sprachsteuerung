# Listen

Listen können beliebige Elemente enthalten, von einfachen Variablen bis hin zu komplexen Strukturen oder Klassen.

Das folgende Beispiel zeigt die Initialisierung einer Liste und das Befüllen mit Werten, z. B. Quadratzahlen.

```python
AList = []
for n in range(10):
    AList.append(n**2)
```

Die Länge der Liste kann mit len(..) überprüft werden:

```python
print('length of the list: ', len(AList))
```

Die Elemente der Liste können z. B. während einer Iteration über die Liste angesprochen werden:

```python
for n in range(len(AList)):
    print('the ', n, '-th element is: ', AList[n])
```

Eine andere Möglichkeit ist der direkte Zugriff auf einzelne Elemente:

```python
for AElement in AList:
    print('An element of the list is: ', AElement)
```
