# Prozeduren
Prozeduren machen den Code nachvollziehbarer.
Im einfachsten Fall ist eine Prozedur eine Folge von Befehlen, die gemeinsam ausgeführt werden:

```python
import numpy as np

def EvaluateSquareRootByNewton(x):
    assert x >= 0, 'input should be greater or equal to zero'
    if x < 1e-16:
        return 0.0
    else:
        y = x / 2
        for n in range(10):
            y = 0.5 * (y + x / y)
        return y

for n in range(10):
    print('square root of ', n, ' = ', EvaluateSquareRootByNewton(n))
```

Analog zu den Einrückungen bei Schleifen und Kontrollstrukturen definiert die Einrückung den Code der Prozedur.
Zu Beginn der Prozedur wird eingerückt.
Am Ende wird die Einrückung wieder reduziert:

```python
def PrintSomething():
    print('Hello')
print('World')

PrintSomething()
```

Die Argumente einer Prozedur sind nicht typsicher, was unerwünschtes Verhalten verursachen kann.

```python
def EvaluateSum(a, b):
    return a + b

print('sum of 3 and 4 = ', EvaluateSum(3, 4))
print('python is not type-safe: ', EvaluateSum('Hello ', 'World'))
```

Argumente können Standardwerte besitzen. Die Reihenfolge muss lauten: zuerst Argumente ohne Standardwerte, dann alle Argumente mit Standardwerten.

```python
def EvaluateDivision(Numerator, Denominator = 1):
    return Numerator / Denominator

assert np.abs(EvaluateDivision(4, 2) - 2) < 1e-10, 'error in EvaluateDivision(4,2)'
print('division of 3 by 4 = ', EvaluateDivision(3, 4))
print('division of 3 by 1 = ', EvaluateDivision(3, 1))
print('division of 3 by default denominator = ', EvaluateDivision(3))
```

Prozeduren sollten so kompakt wie möglich sein:

Sie sollten ohne Scrollen lesbar sein. Im Idealfall gilt das auch für kleine Bildschirme. Dadurch hat der Programmierer eine bessere Chance, die Prozedur mit ihren Eingaben, Algorithmen und Ausgaben zu erfassen.

Wenn du viele Einrückungsebenen hast, solltest du für die inneren Ebenen neue Prozeduren definieren. Das macht Schleifen und Kontrollstrukturen lesbarer.

Im Folgenden wird eine einfache Prozedur gezeigt, die eine Matrixmultiplikation implementiert.
Sie verwendet drei Einrückungsebenen (drei for-Schleifen). In Kombination mit der Einrückung der Prozedur selbst wird der Code recht komplex. Es ist schwer zu erkennen, wofür jede Ebene steht.

```python
def MyMatrixMultiplication(A, B):
    assert A.shape[1] == B.shape[0], 'wrong shapes of input matrices'
    result = np.zeros((A.shape[0], B.shape[1]))
    for row in range(result.shape[0]):
        for col in range(result.shape[1]):
            for n in range(A.shape[1]):
                result[row, col] += A[row, n] * B[n, col]
    return result

A = np.random.randn(7, 5)
B = np.random.randn(5, 6)
C = MyMatrixMultiplication(A, B)
```

Wenn viele Einrückungsebenen nötig sind, lassen sich die inneren Ebenen typischerweise in einer Prozedur kapseln:

```python
def MyDotProduct(x, y):
    assert x.shape[0] == y.shape[0], 'wrong shapes of input vectors'
    result = 0.0
    for n in range(x.shape[0]):
        result += x[n]*y[n]
    return result

def MyMatrixMultiplicationWithLessIndention(A, B):
    
    result = np.zeros((A.shape[0], B.shape[1]))
    for row in range(result.shape[0]):
        for col in range(result.shape[1]):
            result[row, col] = MyDotProduct(A[row, :], B[:, col])
    return result

D = MyMatrixMultiplicationWithLessIndention(A, B)
assert np.sum(np.abs(C- D)) < 1e-10, 'error in evaluation of dot product'
```

## Rekursion
Wenn eine Prozedur sich selbst aufruft, spricht man von Rekursion. Das lässt sich einfach an der Fakultät zeigen:

$N! = \Pi_{n=1}^N n$

```python
def FactorialIterative(N):
    result = 1
    for n in range(2, N+1):
        result *= n
    return result

def FactorialRecursive(N):
    if N < 2:
        return 1
    else:
        return N * FactorialRecursive(N-1)
    
for N in range(10):
    assert FactorialIterative(N) == FactorialRecursive(N), 'iterative and recursive are unequal'
```
