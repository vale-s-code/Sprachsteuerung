# Assertions

Assertions sind eine einfache Möglichkeit, den Code zur Laufzeit zu prüfen.
Eine Assertion besteht aus dem Schlüsselwort assert, gefolgt von einem booleschen Ausdruck und einer Fehlermeldung.
Eine Assertion besteht, wenn der Ausdruck True ergibt.
Die Fehlermeldung wird ausgelöst, falls der Ausdruck False ergibt:

```python
import numpy as np

x = np.random.randn(1)
print('x = ', x)

assert x >= 0.0, 'the square root of a negative number is forbidden for real values'
y = np.sqrt(x)
print('y = sqrt(x) = ', y)
```

## Testcode versus Businesscode
Ein großer Vorteil von Assertions ist:
Sie lassen sich zur Laufzeit deaktivieren. So hat der Entwickler in seiner Version alle Assertions aktiv und kann alle Teile seines Codes prüfen. Für den Kunden kann hingegen im Businesscode die Assertions deaktiviert werden. Dadurch läuft der Code schneller und es treten keine Fehlermeldungen beim Nutzer auf.

## Code Coverage
Code Coverage erklärt, wie viele Zeilen deines Codes durch Assertions ständig geprüft werden. Eine Coverage von 100 % bedeutet, dass der gesamte Businesscode durch Assertions getestet ist.
