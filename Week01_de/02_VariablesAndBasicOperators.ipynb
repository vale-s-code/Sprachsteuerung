# Variablen und Operatoren
In einem Python-Programm kannst du Variablen überall definieren.
Grundlegende Operatoren sind Addition (+), Subtraktion (-), Multiplikation (*) und Division (/).
Um Potenzen auszurechnen, nutzt man $**$.
Zum Beispiel ist das Quadrat von 3: $3**2==9$

```python
a = 7
b = 2
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** b)
```

In einem Jupyter Notebook sind überall deklarierte Variablen in allen Zellen verfügbar.
Zum Beispiel wird die Variable b in der folgenden Codezelle erneut verwendet:

```python
b+4
```

Leider ist Python nicht typsicher.
Im nächsten Beispiel werden zwei Variablen als Ganzzahlen initialisiert und addiert.
Im Anschluss werden beide Variablen als Strings neu definiert und ebenfalls addiert.

```python
a = 3
b = 4
print('integer addition result: ', a + b)

a = 'Hello'
b = ' world'
print('string addition result: ', a + b)
```

Eine Variable um einen bestimmten Wert zu erhöhen, ist eine typische Operation vieler Algorithmen.
Dafür gibt es eine kurze Schreibweise (x += 3):

```python
x = 0
x = x + 3
print(x)
x += 3
print(x)
```
