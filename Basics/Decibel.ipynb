# Decibel
Decibel is a logarithmic measure.

The level $L$ in Decibel of a signal $x(n)$ is evaluated by:

$L=10\log_{10}\left(\frac{1}{N}\sum_{n=0}^{N-1} x^2(n)\right) = 20\log_{10}\left(\frac{1}{N}\sum_{n=0}^{N-1}\left|x(n)\right|\right)$

```python
import numpy as np

def dB(x):
    return 10*np.log10(np.mean(x**2))

gain = np.random.rand(1)*2
x = np.random.randn(100000) * gain 
assert np.abs(dB(x) - 20*np.log10(gain)) < 1e-1, 'error in dB evaluation'
print('All tests passed successfully')
```
