# Delay
The delay between two signals $x(n)$ and $y(n)$ can be evaluated by finding the maximum of the [cross correlation](../Basics/CrossCorrelation.ipynb) between both signals.

```python
import numpy as np

def CrossCorrelation(x, y):
    MaxLength = np.maximum(x.shape[0], y.shape[0])
    FFTLength = 2 ** int(np.ceil(np.log2(2 * MaxLength - 1)))
    X = np.fft.rfft(x, n=FFTLength)
    Y = np.fft.rfft(y, n=FFTLength)
    Phi = np.conj(X) * Y
    phi = np.fft.irfft(Phi)
    return phi

Delay = np.random.randint(0, 200)
x = np.random.randn(1000)
y = np.roll(x, Delay) # simple delay model of rolling the original signal x(n) by Delay samples to get y(n)
phi = CrossCorrelation(x, y)
EstimatedDelay = np.argmax(phi)
assert EstimatedDelay == Delay, 'delay estimation failed'
```
