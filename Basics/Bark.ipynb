# Bark
Humans analyse frequencies in a logarithmic way. The reason for this is the setup of the cochlea: equal spatial distances in the cochlea corresponds to logarithmically spaced resonance frequencies. In cochlea implants $12$ to $22$ equally spaced sensors are used for translating pressure into electrical impulses. From this, it can be argued, that $12$ to $22$ logarithmically spaced bandpass filters are sufficient for speech intelligibility.

There exists two common used logarithmic frequency axis: the mel-scale and the Bark-scale.

Here, the Bark scale is used, instead of the mel-scale. The relationship is: $1$ Bark corresponds to $100$ mel. In the given code, the relationship between frequency $f$ and Bark is given by:

$f = 600\cdot \sinh\left(\frac{\text{Bark}}{6}\right)$

The (logarithmic) frequency resolution of the human ear is roughly $1/27$ Bark as mentioned in <cite>Digitale Sprachsignalverarbeitung from Vary et. al.</cite>.

Human ears can understand frequencies in the range from $16$ Hz up to $15$ kHz.

```python
import numpy as np

### source: https://github.com/TUIlmenauAMS/Python-Audio-Coder/blob/master/PythonPsychoacoustics/psyacmodel.py
def hz2bark(f):
        """ Usage: Bark=hz2bark(f)
            f    : (ndarray)    Array containing frequencies in Hz.
        Returns  :
            Brk  : (ndarray)    Array containing Bark scaled values.
        """
        Brk = 6. * np.arcsinh(f/600.)                                                 
        return Brk

def bark2hz(Brk):
        """ Usage:
        Hz=bark2hs(Brk)
        Args     :
            Brk  : (ndarray)    Array containing Bark scaled values.
        Returns  :
            Fhz  : (ndarray)    Array containing frequencies in Hz.
        """
        Fhz = 600. * np.sinh(Brk/6.)
        return Fhz

r = 48000
f = np.random.rand(1) * r/2
b = hz2bark(f)
assert np.allclose(f, bark2hz(b)), 'hz2bark and bark2hz are not inverses of each other'
```
