# Power

The power of a time continuous signal $x(t)$ is evaluated by:

$P=\frac{1}{T}\int_0^T x^2(t) dt$

In computer based signal processing, the signals are typically time discrete. The power of a digital signal $x(n)$ of length $N$ is defined by:

$P=\frac{1}{N}\sum_{n=0}^{N-1} x^2(n)$

```python
import numpy as np

gain = np.random.rand(1) * 10 
x = np.random.randn(10000) * gain

def EvalPower(x):
    return np.mean(x**2)

P = EvalPower(x)
assert np.sqrt(np.abs(P - gain**2)) < 0.5, f"Expected power {gain**2}, but got {P}"
```
