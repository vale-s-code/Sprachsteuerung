# Standard deviation
The square root of the [variance](./Variance.ipynb) is the standard deviation:

$\sigma = \sqrt{\sigma^2}$

```python
import numpy as np

gain = np.random.rand(1) * 10 
x = np.random.randn(10000) * gain

sigma = np.std(x)

assert np.abs(sigma - gain) < 0.5, f"Expected standard deviation {gain}, but got {sigma}"
```
