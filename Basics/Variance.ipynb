# Variance
The mean of the signal is defined by:

$m=\frac{1}{N}\sum_{n=0}^{N-1} x(n)$

The variance of the signal is defined by:

$\sigma^2=\frac{1}{N}\sum_{n=0}^{N-1} \left(x(n)-m\right)^2$

The following equation is true for the relationship between [power](./Power.ipynb), variance and mean:

$P = \sigma^2 + m^2$

```python
import numpy as np

gain = np.random.rand(1) * 10 
x = np.random.randn(10000) * gain

x += np.random.rand(1) * 10

def EvalVariance(x):
    return np.var(x)

V = EvalVariance(x)
assert np.sqrt(np.abs(V - gain**2)) < 0.5, f"Expected variance {gain**2}, but got {V}"
P =np.mean(x**2)
assert np.sqrt(np.abs(P - (gain**2 + np.mean(x)**2))) < 0.5, f"Expected power {gain**2 + np.mean(x)**2}, but got {P}"    
```
