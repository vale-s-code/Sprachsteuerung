# Pearson correlation
The Pearson correlation is a measure for linear similarity of two signals / two measurements. It is evaluated by

$\varphi=\frac{\sum_n \tilde{x}(n)*\tilde{y}(n)}{\sqrt{\sum_n \tilde{x}^2(n) * \sum_n \tilde{y}^2(n)}}$

with

$\tilde{x}(n) = x(n) - \frac{1}{N}\sum_{n=0}^{N-1}x(n)$

and

$\tilde{y}(n) = y(n) - \frac{1}{N}\sum_{n=0}^{N-1}y(n)$

```python
import numpy as np

def PearsonCorrelation(x, y):
    x_zeromean = x - np.mean(x)
    y_zeromean = y - np.mean(y)
    return np.sum(x_zeromean*y_zeromean) / np.sqrt(np.sum(x_zeromean**2)*np.sum(y_zeromean**2))

x = np.random.rand(1000)
y = np.random.rand(1000)
phi = PearsonCorrelation(x, y)
assert np.abs(phi) <= 1.0, 'Pearson correlation is restricted to -1..1'
y = np.random.rand(1) * x
phi = PearsonCorrelation(x, y)
assert np.abs(phi - 1.0) < 1e-6, 'Pearson correlation of perfectly correlated data should be 1'
y = -np.random.rand(1) * x
phi = PearsonCorrelation(x, y)
assert np.abs(phi + 1.0) < 1e-6, 'Pearson correlation of perfectly anti-correlated data should be -1'
print('All tests passed successfully')
```
