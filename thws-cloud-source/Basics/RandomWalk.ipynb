# Random Walk
Random walk is an optimization algorithm which minimizes a given cost function. The advantages of this algorithm are:

- The implementation of this algorithm is very simple.
- The evaluation of the gradient is not necessary.

The random walk starts at a random position. The cost function at this position is evaluated. From this current position the cost function at a random neighbouring position is evaluated. If this random neighbouring position leads to a smaller cost function, this new position is kept as the starting position of the next iteration.

```python
import numpy as np

def CostFunction(x):
    return x**2 - 4*x + 5

BestPosition = np.random.randn(1)
BestCost = CostFunction(BestPosition)
for n in range(1000):
    NewPosition = BestPosition + np.random.randn(1)
    NewCost = CostFunction(NewPosition)
    if NewCost < BestCost:
        BestPosition = NewPosition
        BestCost = NewCost
        print(f"New best position: {BestPosition}, New best cost: {BestCost}")    
```
