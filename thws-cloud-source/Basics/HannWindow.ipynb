# Hann window
The Hann window is defined by

$w(n)=\frac{1}{2}\left(1+\cos\left(2\pi n / N \right)\right)\cdot\text{rect}\left(\frac{n}{N}\right)$.

The Hann window is restricted to the range $-\frac{N}{2}< n < \frac{N}{2}$ by the rect-Signal:

$\text{rect}(t)= 
\begin{cases}
    1,& \text{if } -0.5<t<0.5\\
    0,& \text{otherwise}
\end{cases}
$

The Hann window is a standard window function for utilizing local signal analysis with the DFT or the STFT.

```python
import numpy as np
import matplotlib.pyplot as plt

def HannWindow(N):
    n = np.arange(N)
    return 0.5 * (1 - np.cos(2*np.pi*(n+0.5)/N))

w = HannWindow(51)
plt.plot(w)
plt.xlabel('n')
plt.ylabel('w[n]')
plt.title('Hann window')
plt.show()
```
