## Kernideen fuer die Probeklausur

Die Probeklausur prueft nicht nur einzelne Definitionen, sondern vor allem die Verbindung zwischen Konzepten und praktischer Umsetzung:

- Ein Audiosignal ist eine Folge diskreter Samples, wenn es digital verarbeitet wird.
- Vorverarbeitung veraendert Signale so, dass relevante Informationen leichter analysiert werden koennen.
- Frequenzanalyse zeigt, welche Frequenzanteile in einem Signal enthalten sind.
- Die STFT verbindet Zeit- und Frequenzinformation, indem sie kurze Signalabschnitte analysiert.
- Filter reduzieren oder verstaerken bestimmte Frequenzbereiche.
- Entrauschung ist immer ein Kompromiss zwischen Rauschunterdrueckung und Erhalt des Nutzsignals.
- Python-Code in der Klausur sollte zuerst in Datenformen, Achsen und Signalbedeutung zerlegt werden.

## Wichtige Formeln

Abtastzeit und Abtastfrequenz:

$T_s = \frac{1}{f_s}$

Nyquist-Grenze:

$f_\text{max} = \frac{f_s}{2}$

Diskrete Fourier-Transformation:

$X(k)=\sum_{n=0}^{N-1}x(n)e^{-j2\pi kn/N}$

Faltung fuer FIR-Filter:

$y(n)=\sum_{m=0}^{M-1}h(m)\cdot x(n-m)$

Additives Rauschmodell:

$y(n)=x(n)+d(n)$

Einfache Wiener-Filter-Idee:

$H(f)=\frac{|X(f)|^2}{|X(f)|^2+|D(f)|^2}$

Signal-Rausch-Verhaeltnis:

$SNR = 10\log_{10}\left(\frac{P_\text{Signal}}{P_\text{Rauschen}}\right)$

## Typische Fehler

- `fs` als Signallaenge statt als Abtastfrequenz interpretieren.
- Sekunden und Samples in derselben Rechnung vermischen.
- Die Nyquist-Grenze mit der Abtastfrequenz verwechseln.
- Betragsspektrum, Leistungsspektrum und komplexes Spektrum gleich behandeln.
- Bei der STFT Zeitachse und Frequenzachse vertauschen.
- Filterkoeffizienten als Frequenzen interpretieren.
- Rauschunterdrueckung nur nach Zahlen bewerten und hoerbare Artefakte ignorieren.
- Code auswendig nachbauen, ohne Arrayformen und Achsen zu pruefen.
