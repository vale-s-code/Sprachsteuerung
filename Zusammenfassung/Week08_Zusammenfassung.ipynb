## Kernidee

Ein Nutzsignal von einer Schallquelle kommt an mehreren Mikrofonen in aehnlicher Form an, aber mit unterschiedlichen Laufzeiten. Das Rauschen der Mikrofone wird als weitgehend unkorreliert angenommen. Wenn die Mikrofonsignale zuerst passend verschoben und dann gemittelt werden, addiert sich das Nutzsignal konstruktiv, waehrend das Rauschen weniger stark waechst.

Der Ablauf ist:

1. Aufnahme mit mehreren Mikrofonen.
2. Schaetzung der relativen Verzoegerung ueber Kreuzkorrelation.
3. Zeitliche Ausrichtung der Signale.
4. Mittelung beziehungsweise Summation der ausgerichteten Signale.
5. Bewertung ueber SNR, Korrelation und Hoereindruck.

## Wichtige Formeln

Abstand zwischen Mikrofonposition $m$ und Schallquellenposition $s$:

$\text{dist}=\sqrt{(m_x-s_x)^2+(m_y-s_y)^2}$

Laufzeit und Sample-Verzoegerung:

$T_\text{seconds}=\frac{\text{dist}}{c}$

$T=\text{round}(T_\text{seconds}\cdot r)$

Kreuzkorrelation zur Verzoegerungsschaetzung:

$\varphi(m)=\sum_{n=0}^{N-1} y_1(n)\cdot y_2(n+m)$

SNR-Gewinn bei zwei Mikrofonen mit perfekter Ausrichtung und unkorreliertem Rauschen:

$10\log_{10}\left(\frac{4}{2}\right)\approx 3.01\,\text{dB}$

Normierte Korrelation:

$\varphi_\text{norm}=\frac{\max_m |\varphi(m)|}{\sqrt{\sum_n s^2(n)\cdot\sum_n g^2(n)}}$

```python
import numpy as np

c = 343      # m/s
r = 48000    # Hz
distance_difference = 0.2 * np.sin(np.deg2rad(30))
delay_seconds = distance_difference / c
delay_samples = round(delay_seconds * r)

delay_seconds, delay_samples
```

Das Beispiel zeigt die typische Umrechnung von Geometrie ueber Laufzeit in eine ganzzahlige Sample-Verzoegerung.

## Typische Fehler

- Signale addieren, bevor die Laufzeitverzoegerung kompensiert wurde.
- Das Vorzeichen der Verzoegerung nicht an die gewaehlte Referenz anpassen.
- Korrelation mit Kausalitaet verwechseln.
- Bei unkorreliertem Rauschen Amplituden statt Leistungen addieren.
- Den SNR-Gewinn von 3 dB fuer zwei Mikrofone auf beliebige Situationen uebertragen, obwohl gleiche Rauschpegel und perfekte Ausrichtung vorausgesetzt werden.
- Fernfeld- und Nahfeldannahmen vermischen.
- Winkelangaben in Grad direkt in trigonometrische Funktionen einsetzen, die Bogenmass erwarten.
