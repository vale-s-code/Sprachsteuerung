## FIR-Filter

Ein FIR-Filter (Finite Impulse Response) besitzt eine endliche Impulsantwort $h(n)$. Das Ausgangssignal entsteht durch Faltung:

$y(n)=x(n)*h(n)=\sum_{m=0}^{M-1}h(m)\cdot x(n-m)$

Ein FIR-Filter bildet also jedes neue Ausgangssample aus aktuellen und vergangenen Eingangssamples.

Vorteil: FIR-Filter sind bei endlicher Impulsantwort stabil. Nachteil: Fuer scharfe Filter braucht man oft viele Koeffizienten.

```python
import numpy as np

x = np.array([0, 1, 2, 0])
h = np.array([1, 2, 1])
y = np.convolve(x, h)
y
```

## Faltung einfach erklaert

Bei der Faltung wird die Impulsantwort ueber das Signal geschoben. An jeder Position wird gewichtet summiert.

Wenn $h(n)$ zum Beispiel Werte mittelt, glaettet der Filter das Signal. Wenn $h(n)$ schnelle Aenderungen betont, wirkt der Filter eher wie ein Hochpass.

## Filtertypen

- Tiefpass: laesst niedrige Frequenzen durch und daempft hohe Frequenzen.
- Hochpass: laesst hohe Frequenzen durch und daempft niedrige Frequenzen.
- Bandpass: laesst einen mittleren Frequenzbereich durch und daempft tiefe und hohe Frequenzen.
- Bandsperre: daempft einen bestimmten Bereich und laesst den Rest durch.

Fuer Sprache ist ein Bandpass oft sinnvoll, weil viele Stoerungen ausserhalb des Sprachbereichs liegen.

## Bandpass fuer Sprachsignale

Typische Sprachbereiche:

- Narrowband: ca. 300 Hz bis 3400 Hz
- Wideband: ca. 50 Hz bis 7000 Hz
- Fullband: ca. 20 Hz bis 20000 Hz

Wenn ein Sensor breitbandiges Rauschen misst, aber das Nutzsignal Sprache ist, kann ein Bandpass ausserhalb des Sprachbereichs Rauschleistung entfernen. Dadurch steigt das SNR, falls das Nutzsignal durch den Filter kaum geschwaecht wird.

## FIR im Frequenzbereich

Faltung im Zeitbereich entspricht Multiplikation im Frequenzbereich:

$y(n)=x(n)*h(n)$

entspricht:

$Y(f)=X(f)\cdot H(f)$

Das ist wichtig, weil Filter im Frequenzbereich oft anschaulicher sind: $H(f)$ zeigt, welche Frequenzen durchgelassen oder gedaempft werden.

```python
# Betrag der Uebertragungsfunktion eines einfachen Glattungsfilters.
h = np.array([1, 2, 1]) / 4
H = np.fft.rfft(h, n=128)
frequencies = np.linspace(0, 0.5, len(H))

frequencies[:5], np.abs(H[:5])
```

## Symmetrische FIR-Filter und Verzoegerung

Ein FIR-Filter ist symmetrisch, wenn gilt:

$h(n)=h(N-n-1)$

Symmetrische FIR-Filter haben eine lineare Phase. Das bedeutet praktisch: Frequenzanteile im Durchlassbereich werden gleichmaessig verzoegert.

Die Verzoegerung betraegt ungefaehr:

$D=\frac{N}{2}$

oder genauer bei ungerader Laenge oft $(N-1)/2$ Samples, je nach Definition im Notebook.

## Amplitudenanpassung nach Filterung

Ein Filter kann nicht nur verzoegern, sondern auch die Amplitude veraendern. Deshalb wird im Notebook ein Faktor $a$ gesucht, der das gefilterte Signal bestmoeglich an das Vergleichssignal anpasst.

Die Idee ist, den quadratischen Fehler zu minimieren:

$\sum_n \left(a\cdot y(n-D)-z(n)\right)^2$

Das ist eine einfache Form der optimalen Skalierung.

## IIR-Filter

Ein IIR-Filter (Infinite Impulse Response) nutzt neben Eingangswerten auch vergangene Ausgangswerte:

$y(n)=\sum_{k=0}^{K}b_kx(n-k)-\sum_{l=1}^{L}a_ly(n-l)$

Dadurch kann ein IIR-Filter mit wenigen Koeffizienten stark filtern. Gleichzeitig kann er instabil werden, weil alte Ausgangswerte immer wieder zurueckgefuehrt werden.

## z-Transformation und Uebertragungsfunktion

Die Uebertragungsfunktion eines IIR-Filters kann mit der z-Transformation beschrieben werden:

$H(z)=\frac{Y(z)}{X(z)}$

Fuer die Frequenzantwort setzt man:

$z=e^{j2\pi f/r}$

Dann erhaelt man $H(f)$ und kann sehen, ob der Filter wie Tiefpass, Bandpass oder Hochpass wirkt.

## Pole, Nullstellen und Stabilitaet

Pole und Nullstellen beschreiben die Uebertragungsfunktion im z-Bereich.

- Nullstellen daempfen bestimmte Frequenzbereiche.
- Pole verstaerken bestimmte Bereiche.
- Ein IIR-Filter ist stabil, wenn alle Pole innerhalb des Einheitskreises liegen.

Das Pol-Nullstellen-Diagramm ist daher ein wichtiges Werkzeug zur Stabilitaetsanalyse.

## Typische Fehler

- FIR und IIR nur ueber Namen unterscheiden, aber die Rueckkopplung beim IIR vergessen.
- Grenzfrequenz und Mittenfrequenz beim Bandpass verwechseln.
- Filterverzoegerung ignorieren und Signale falsch vergleichen.
- Betrag in dB und linearen Betrag vermischen.
- IIR-Stabilitaet nicht pruefen.
