# Woche 3: Diskrete Fourier-Transformation

Die diskrete Fourier-Transformation (DFT) ist eines der wichtigsten Werkzeuge der digitalen Signalverarbeitung. Sie zeigt, welche Frequenzen in einem zeitdiskreten Signal enthalten sind.

## Zeitbereich und Frequenzbereich

Im Zeitbereich betrachtet man die Samples $x(n)$ direkt. Man sieht, wie sich das Signal ueber die Zeit veraendert.

Im Frequenzbereich betrachtet man $X(k)$. Man sieht, wie stark bestimmte Frequenzen im Signal enthalten sind.

Die DFT ist die Bruecke vom Zeitbereich in den Frequenzbereich.

## DFT-Formel

Die DFT eines Signals $x(n)$ lautet:

$X(k)=\sum_{n=0}^{N-1} x(n)\cdot e^{-j2\pi nk/K}$

Bedeutung:

- $n$: Zeitindex
- $k$: Frequenzindex
- $N$: Anzahl betrachteter Samples
- $K$: Transformationslaenge
- $X(k)$: komplexer Frequenzwert

Der Betrag $|X(k)|$ sagt, wie stark die Frequenzkomponente ist. Die Phase sagt, wie die Schwingung zeitlich verschoben ist.

```python
import numpy as np

r = 1000
N = 1000
f_signal = 100
n = np.arange(N)
x = np.sin(2 * np.pi * f_signal * n / r)

X = np.fft.fft(x)
frequencies = np.arange(N) * r / N
peak_frequency = frequencies[np.argmax(np.abs(X[:N // 2]))]

peak_frequency
```

## Frequenzaufloesung

Die Frequenzaufloesung beschreibt den Abstand zwischen zwei Frequenzbins:

$\Delta f = \frac{r}{K}$

- groesseres $K$ fuehrt zu kleinerem $\Delta f$
- kleineres $\Delta f$ bedeutet feinere Frequenzachse

Wichtig: Eine feinere Frequenzachse bedeutet nicht automatisch mehr echte Information. Die Information kommt aus der Laenge des beobachteten Signals.

```python
r = 1000
for K in [100, 500, 1000, 2000]:
    print(K, '-> Delta f =', r / K, 'Hz')
```

## Zero Padding

Zero Padding bedeutet, dass ein Signal vor der DFT mit Nullen verlaengert wird. Dadurch wird $K$ groesser und die Frequenzachse feiner abgetastet.

Aber: Zero Padding erzeugt keine neue Messinformation. Es interpoliert das Spektrum nur feiner.

Praktisch hilft Zero Padding, Peaks genauer zu lokalisieren oder Spektren glatter aussehen zu lassen.

```python
r = 1000
N = 100
f_signal = 123
n = np.arange(N)
x = np.sin(2 * np.pi * f_signal * n / r)

for K in [N, 10 * N]:
    X = np.fft.fft(x, n=K)
    f = np.arange(K) * r / K
    peak = f[np.argmax(np.abs(X[:K // 2]))]
    print('K =', K, 'Peak bei ca.', peak, 'Hz')
```

## Reelle Signale und Symmetrie

Audiosignale sind reellwertig. Bei reellen Eingangssignalen ist das DFT-Spektrum symmetrisch. Deshalb betrachtet man oft nur die Frequenzen von 0 Hz bis zur Nyquist-Frequenz.

In NumPy ist dafuer `np.fft.rfft` praktisch. Es berechnet nur den positiven Frequenzanteil fuer reelle Signale.

```python
x = np.random.randn(16)
X_full = np.fft.fft(x)
X_positive = np.fft.rfft(x)

len(X_full), len(X_positive)
```

## Betrag und Phase

Das DFT-Ergebnis ist komplex. Deshalb gibt es zwei wichtige Darstellungen:

- Betragsspektrum: `np.abs(X)`
- Phasenspektrum: `np.angle(X)`

Fuer viele Aufgaben in der Vorlesung ist der Betrag zuerst wichtiger, weil er zeigt, welche Frequenzen stark enthalten sind.

## Typische Fehler

- Frequenzindex $k$ mit physikalischer Frequenz in Hertz verwechseln.
- Nur den groessten Peak betrachten und die Spiegelung nicht verstehen.
- Zero Padding als echte Erhoehung der Information interpretieren.
- Betrag und komplexen DFT-Wert verwechseln.
- Abtastrate bei der Frequenzachse vergessen.
