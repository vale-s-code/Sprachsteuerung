## Analoges und digitales Signal

Ein analoges Signal $x(t)$ ist kontinuierlich in der Zeit. Es kann theoretisch zu jedem Zeitpunkt einen Wert haben.

Ein digitales Signal $x(n)$ besteht aus einzelnen Samples. Der Index $n$ zaehlt die Samples. Der Zusammenhang zwischen Zeit und Sampleindex ist:

$t = n \cdot T$

mit:

- $T$: Abtastperiode in Sekunden
- $r$: Abtastrate in Samples pro Sekunde
- $T = \frac{1}{r}$

```python
import numpy as np

r = 1000          # Abtastrate in Hz
T = 1 / r        # Abtastperiode
N = 10           # Anzahl Samples
n = np.arange(N)
t = n * T

t
```

## Abtastrate

Die Abtastrate $r$ gibt an, wie viele Samples pro Sekunde aufgenommen werden.

Beispiele:

- Telefonie Narrowband: haeufig 8 kHz
- Audio-CD: 44.1 kHz
- Viele Sprachsignale im Kurs: 48 kHz

Eine hoeherer Abtastrate kann hoehere Frequenzen darstellen, erzeugt aber auch mehr Daten.

## Abtasttheorem

Das Abtasttheorem sagt:

Ein Signal kann eindeutig rekonstruiert werden, wenn die Abtastrate groesser als das Doppelte der hoechsten enthaltenen Frequenz ist.

$r > 2 \cdot f_\text{max}$

Die halbe Abtastrate heisst Nyquist-Frequenz:

$f_\text{Nyquist}=\frac{r}{2}$

Frequenzen oberhalb der Nyquist-Frequenz koennen nicht korrekt dargestellt werden.

```python
r = 48000
f_nyquist = r / 2
f_nyquist
```

## Aliasing

Aliasing entsteht, wenn ein Signal Frequenzen enthaelt, die hoeher als die Nyquist-Frequenz sind. Diese Frequenzen erscheinen dann als falsche niedrigere Frequenzen im digitalen Signal.

Einfach gesagt: Das digitale System verwechselt eine zu hohe Frequenz mit einer anderen Frequenz.

Deshalb verwendet man vor der Abtastung oft einen Anti-Aliasing-Tiefpass. Er entfernt Frequenzen oberhalb der Nyquist-Frequenz.

```python
# Beispiel: Frequenz 700 Hz bei Abtastrate 1000 Hz.
# Nyquist ist 500 Hz, also kann 700 Hz nicht korrekt dargestellt werden.
r = 1000
f_original = 700
f_alias = abs(f_original - r)

f_alias
```

## Sinus als Eingangssignal

Viele Grundlagen werden mit Sinus-Signalen erklaert, weil ein Sinus genau eine Frequenz besitzt.

Ein Sinus im zeitdiskreten Bereich kann so geschrieben werden:

$x(n)=\sin\left(2\pi f \frac{n}{r}\right)$

Dabei ist $f$ die Frequenz in Hertz und $r$ die Abtastrate.

```python
r = 1000
f = 100
N = 20
n = np.arange(N)
x = np.sin(2 * np.pi * f * n / r)

x[:10]
```

## Quantisierung

Beim Digitalisieren wird nicht nur die Zeit diskretisiert, sondern auch die Amplitude. Ein ADC kann nur endlich viele Werte darstellen.

Beispiel:

- 8 Bit: 256 moegliche Werte
- 16 Bit: 65536 moegliche Werte

Quantisierung erzeugt einen kleinen Fehler, weil reale Messwerte auf darstellbare Werte gerundet werden.

```python
# Sehr einfache Quantisierung auf 8 Stufen zwischen -1 und 1.
x = np.linspace(-1, 1, 11)
stufen = 8
x_quantisiert = np.round((x + 1) / 2 * (stufen - 1)) / (stufen - 1) * 2 - 1

x, x_quantisiert
```

## Signalenergie und Lautstaerkeidee

Ein Signal mit groesseren Amplituden hat meist mehr Leistung/Energie. Viele spaetere Verfahren vergleichen Signalanteile und Rauschanteile ueber Quadrate der Amplituden.

Deshalb tauchen Ausdruecke wie $x(n)^2$, Mittelwert von Quadraten oder Root Mean Square spaeter sehr haeufig auf.

## Typische Fehler

- Abtastrate und Frequenzachse verwechseln.
- Hertz und Sampleindex vermischen.
- Nyquist-Frequenz vergessen.
- Aliasing als Programmierfehler interpretieren, obwohl es ein mathematischer Effekt ist.
- Zu wenige Samples erzeugen und dann ein Signal falsch beurteilen.
