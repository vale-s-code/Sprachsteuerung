## Variablen und Datentypen

Eine Variable ist ein Name fuer einen Wert. Python merkt sich den Typ automatisch.

Typische Datentypen:

- `int`: ganze Zahl, zum Beispiel `5`
- `float`: Kommazahl, zum Beispiel `5.0`
- `str`: Text, zum Beispiel `'Hallo'`
- `bool`: Wahrheitswert, also `True` oder `False`
- `list`: Liste mit mehreren Elementen

Wichtig fuer Signalverarbeitung: Signale werden spaeter meistens als `numpy`-Arrays gespeichert. Ein Array enthaelt viele Zahlenwerte, zum Beispiel alle Samples eines Audiosignals.

````python
zahl = 5
kommazahl = 5.0
text = 'Signalverarbeitung'
ist_groesser = zahl > 3
liste = [1, 2, 3]

type(zahl), type(kommazahl), type(text), type(ist_groesser), type(liste)
````

## Operatoren

Operatoren fuehren Rechnungen oder Vergleiche aus.

- `+`, `-`, `*`, `/`: Grundrechenarten
- `**`: Potenz, zum Beispiel `2**3 = 8`
- `%`: Rest einer Division
- `==`: Gleichheit pruefen
- `!=`: Ungleichheit pruefen
- `<`, `<=`, `>`, `>=`: Vergleiche

Achtung: `=` weist einen Wert zu, `==` vergleicht zwei Werte.

````python
a = 10
b = 3

rechnung = a / b
potenz = b ** 2
vergleich = a == b

rechnung, potenz, vergleich
````

## Bedingungen

Mit Bedingungen entscheidet das Programm, welcher Code ausgefuehrt wird.

Typische Struktur:

```python
if bedingung:
    ...
elif andere_bedingung:
    ...
else:
    ...
```

Die Einrueckung ist in Python Teil der Sprache. Falsche Einrueckung ist eine haeufige Fehlerquelle.

````python
wert = 7

if wert < 0:
    ergebnis = 'negativ'
elif wert == 0:
    ergebnis = 'null'
else:
    ergebnis = 'positiv'

ergebnis
````

## Schleifen

Schleifen wiederholen Code.

Eine `for`-Schleife wird verwendet, wenn bekannt ist, welche Elemente durchlaufen werden sollen. Eine `while`-Schleife wird verwendet, solange eine Bedingung wahr ist.

In den Signalverarbeitungs-Notebooks werden Schleifen oft benutzt, um Samples, Frequenzbins oder Spektrogrammspalten zu verarbeiten.

````python
werte = [1, 2, 3, 4]
summe = 0

for wert in werte:
    summe += wert

summe
````

## Funktionen

Funktionen machen Code wiederverwendbar. Eine Funktion bekommt Eingaben, verarbeitet sie und gibt ein Ergebnis zurueck.

Begriffe:

- Parameter: Name in der Funktionsdefinition
- Argument: konkreter Wert beim Aufruf
- Rueckgabewert: Ergebnis nach `return`

Funktionen sind wichtig, weil spaetere Notebooks viele Verarbeitungsschritte kapseln, zum Beispiel Fensterfunktionen, Filter oder Spektrogrammberechnung.

````python
def mittelwert(x):
    return sum(x) / len(x)

werte = [2, 4, 6, 8]
mittelwert(werte)
````

## Listen und NumPy-Arrays

Python-Listen sind allgemein. NumPy-Arrays sind fuer numerische Rechnungen optimiert.

Der wichtigste Unterschied: Bei NumPy wirken Operationen oft elementweise auf alle Werte.

````python
import numpy as np

liste = [1, 2, 3]
array = np.array([1, 2, 3])

# Liste + Liste haengt aneinander.
liste_ergebnis = liste + liste

# Array + Array addiert elementweise.
array_ergebnis = array + array

liste_ergebnis, array_ergebnis
````

## Klassen kurz erklaert

Eine Klasse ist ein Bauplan fuer Objekte. Ein Objekt kann Daten speichern und Methoden besitzen.

In den spaeteren Notebooks wird zum Beispiel eine Spektrogramm-Klasse verwendet. Sie speichert Parameter wie Abtastrate, Fensterlaenge und Hopgroesse und bietet Methoden zur STFT/ISTFT an.

````python
class SignalInfo:
    def __init__(self, sampling_rate):
        self.sampling_rate = sampling_rate

    def nyquist_frequency(self):
        return self.sampling_rate / 2

info = SignalInfo(48000)
info.nyquist_frequency()
````

## Assertions und Fehlersuche

`assert` prueft eine Annahme. Wenn die Annahme falsch ist, stoppt Python mit einem Fehler.

Das ist in Uebungsnotebooks nuetzlich, weil automatisch getestet werden kann, ob eine Aufgabe richtig geloest wurde.

````python
def positive_wurzel(x):
    assert x >= 0, 'x muss positiv sein'
    return np.sqrt(x)

positive_wurzel(9)
````

## Typische Fehler

- Zelle nicht ausgefuehrt: Variable oder Funktion ist unbekannt.
- Falsche Reihenfolge: Eine spaetere Zelle haengt von einer frueheren ab.
- Einrueckung falsch: Python erkennt den Block nicht.
- `=` statt `==`: Zuweisung und Vergleich verwechselt.
- Liste statt NumPy-Array: Rechenoperation verhaelt sich anders.
- Pfade stimmen nicht: Datei wird nicht gefunden.
